```python
def realisedVol(windowLength):
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.window.rolling.Rolling.std.html
    return df.iloc[:,2].rolling(windowLength).std()
```

```python
def simplePriceMovingAvg(maWindow):
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.window.rolling.Rolling.std.html
    return df.iloc[:,1].rolling(maWindow).mean()
```

```python
def addColumnandTitle(df1,df2, title):
    #df = pd.concat([df,simpleMa], 1)
    df3 = pd.concat([df1, df2], 1)
    names = list(df3.columns)
    names[-1] = title
    df3.columns = names
    return df3
```

```python
titles = ['Long returns','{}-day MA'.format(maWindow), '{}-day Historical vol'.format(windowLength)]

returns = df.iloc[:,1].pct_change()
df = addColumnandTitle(df, returns, titles[0])

maWindow = 15
simpleMa = simplePriceMovingAvg(maWindow)

windowLength = 10
vol = realisedVol(windowLength)

df = addColumnandTitle(df,simpleMa, titles[1])
df = addColumnandTitle(df,vol, titles[2])
df
```
